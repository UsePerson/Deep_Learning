```python
import pandas as pd
import numpy as np
from NN import nn
```

```python
train_data  = np.loadtxt('./train/train_img.txt', delimiter=',')
train_label = np.loadtxt('./train/train_label.txt', delimiter=',')
train_data  = np.array(train_data)
train_label = pd.get_dummies(train_label)

train = np.append(train_data, np.array(train_label), axis=1)
# np.random.shuffle(train)
```

```python
NN = nn(ef='cross_entropy', lr=0.001)
NN.add_layer(np.size(train_data, 1), 128)
NN.add_layer(128, 3)
```

```python
NN.SGD(train, 0.25, 100)
```

```python
test_data = np.loadtxt('./test/test_img.txt', delimiter=',')
test_label = np.loadtxt('./test/test_label.txt', delimiter=',')

test_data = np.array(test_data)
test_label = pd.get_dummies(test_label)

test = np.append(test_data, np.array(test_label), axis=1)
```

```python
NN.test(test)
```
